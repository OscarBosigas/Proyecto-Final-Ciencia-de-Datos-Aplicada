```python
#tabla de frecuencias 

#Se realiza tanto para el cultivo, como para el departamento esto con tal de determinar
#en que zonas y que cultivos tienen mayor cantidad de datos para ser estudiados

#Se define una funcion que nos permita realizar este cálculo para cualquier tipo de campo 
#que se quiera analizar respecto a su frecuencia relativa

def  rel_freq (x): 
    freqs = [(value, x.count (value) / len (x)) for value in set (x)] 
    return freqs
```

```python

def  rel_freq (x): 
    freqs = [(value, x.count (value) / len (x)) for value in set (x)] 
    return freqs
```

```python
freq=rel_freq(list(Cosechas_1['DEPARTAMENTO']))
freq_tabla=pd.DataFrame(freq)
freq_tabla=freq_tabla.set_axis(['Departamento','Frecuencia relativa'],axis=1)
freq_tabla['Frecuencia relativa %']=freq_tabla['Frecuencia relativa'].apply(lambda x:format(x,'.2%'))
Mayor_frecuencia_departamentos = freq_tabla.sort_values('Frecuencia relativa',ascending=False)
```

```python
#Detallamos de mayor a menor los departamentos donde mas se conecentra la recolecta de informacion 
Mayor_frecuencia_departamentos
```

```python
freq=rel_freq(list(Cosechas_1['CULTIVO']))
freq_tabla=pd.DataFrame(freq)
freq_tabla=freq_tabla.set_axis(['CULTIVO','Frecuencia relativa'],axis=1)
freq_tabla['Frecuencia relativa %']=freq_tabla['Frecuencia relativa'].apply(lambda x:format(x,'.5%'))
Mayor_frecuencia_cultivo = freq_tabla.sort_values('Frecuencia relativa',ascending=False)
```

```python
#Detallamos de mayor a menor los departamentos donde mas se conecentra la recolecta de informacion
Mayor_frecuencia_cultivo.tail(50)
```
